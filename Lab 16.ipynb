```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import pandas as pd
```

```python
x = 10 ** np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})
```

```python
grid = sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={"s": 100})
```

```python
grid.set_xticklabels(rotation=30)

plt.show()
```
